## 5. Data Cleaning 

```python
cat_cols = ['Type of Food','Event Type','Storage Conditions','Purchase History',
            'Seasonality','Preparation Method','Geographical Location','Pricing']

for c in cat_cols:
    food_df[c] = food_df[c].astype(str).str.strip().str.title()
```

```python
food_df
```

```python
food_df['Number of Guests'] = pd.to_numeric(food_df['Number of Guests'], errors='coerce')
food_df['Quantity of Food'] = pd.to_numeric(food_df['Quantity of Food'], errors='coerce')
food_df['Wastage Food Amount'] = pd.to_numeric(food_df['Wastage Food Amount'], errors='coerce')
```

```python
food_df
```

```python
food_df.isnull().sum()
```

```python
print("duplicates:", food_df.duplicated().sum())
```

## 6. Creating new columns from old one's - Feature Engineering 

```python
food_df['percent_wasted'] = np.where(food_df['Quantity of Food']>0,
                                food_df['Wastage Food Amount'] / food_df['Quantity of Food'] * 100,
                                np.nan)
food_df['wastage_per_guest'] = np.where(food_df['Number of Guests']>0,
                                   food_df['Wastage Food Amount'] / food_df['Number of Guests'],
                                   np.nan)
food_df['quantity_per_guest'] = np.where(food_df['Number of Guests']>0,
                                    food_df['Quantity of Food'] / food_df['Number of Guests'],
                                    np.nan)

food_df[['Wastage Food Amount','Quantity of Food','percent_wasted','wastage_per_guest']].describe().T
```

```python
food_df
```

```python
plt.figure(figsize=(8,6))
corr = food_df[['Number of Guests','Quantity of Food','Wastage Food Amount','percent_wasted','wastage_per_guest','quantity_per_guest']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features', fontsize=14)
plt.show()
```

The total amount of food prepared is also strongly related to the number of guests. More guests mean more food was prepared.
However, a weak relationship was observed between the number of guests and the food waste per person. This suggests that simply having a bigger or smaller event doesn't necessarily mean each individual person will waste more or less.

```python
plt.figure(figsize=(7,5))
sns.barplot(x='Storage Conditions', y='percent_wasted', data=food_df, palette='Set2', ci=None)
plt.title('Average Percentage Wastage by Storage Conditions')
plt.xlabel('Storage Conditions')
plt.ylabel('Average % Wasted')
plt.show()
```

### From the above graph we can say that the average percentage wastage of food observed is nearly equal for both Refrigerated and room temperature storage condidtions. 

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Type of Food', y='percent_wasted', data=food_df, palette='pastel')
plt.title('Percentage Wastage Distribution by Type of Food')
plt.xlabel('Type of Food')
plt.ylabel('Percentage Wasted')
plt.xticks(rotation=45)
plt.show()
```

- The median percentage wastage is highest for Fruits and Baked Goods, indicating these types of food are typically wasted more than others.

- Fruits show the highest overall maximum wastage, with an extreme outlier at nearly 16%. This suggests there are some cases of very high wastage for fruits.

- The percentage of wastage for Dairy Products is relatively consistent, with a smaller spread of data points. The range of wastage for dairy products is from about 4% to 8%. 

- The percenatge wastage for Meat and Fruits has a wider spread than the other food types, hence, there is more variability observed in their wastage percentages.

### 3. Multivariate Analysis : 

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x='Seasonality', 
    y='percent_wasted', 
    hue='Storage Conditions',
    data=food_df, 
    ci=None, 
    palette='Set2'
)
plt.title('Average Percentage Wastage by Seasonality and Storage Type')
plt.xlabel('Seasonality')
plt.ylabel('Average % Wasted')
plt.legend(title='Storage Conditions')
plt.show()
```

The average percentage of food wasted is very similar for both refrigerated and room-temperature items. This trend holds true across all seasons, suggesting that the type of storage has very little impact on the average percentage of food wasted.

```python
plt.figure(figsize=(9,5))
sns.barplot(
    x='Event Type', y='percent_wasted', hue='Seasonality',
    data=food_df, ci=None, palette='Set2'
)
plt.title('Average Percentage Wasted by Event Type Across Seasons')
plt.xlabel('Event Type')
plt.ylabel('Average % Wasted')
plt.xticks(rotation=45)
plt.legend(title='Seasonality')
plt.show()
```

### the average percentage of food wastage observed is highest in the Summer season across all the Event types followed by winter and lastly, all-seasons seasonality

The analysis provides a view into the company’s food wastage patterns. The food company should focus on corporate events in suburban areas, with a special focus on managing fruits and baked goods. While changing storage conditions might not have a major impact on waste, understanding the relationships between factors like the number of guests and food quantity can help the food company plan more effectively and achieve the goal of preventing food shortages.
